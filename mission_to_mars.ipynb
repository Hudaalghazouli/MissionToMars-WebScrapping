```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import requests
import os
import pandas as pd
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://redplanetscience.com/'
browser.visit(url)
browser.is_element_present_by_css('div.list_text', wait_time=1)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')

print(soup.prettify())
```

```python
last_new = soup.select_one('div.list_text')
```

```python
last_new.find('div',class_='content_title')
```

```python
last_new_text= last_new.find('div',class_='content_title').text
last_new_text
```

```python
last_new_paragraph = last_new.find('div',class_='article_teaser_body').text
last_new_paragraph
```

```python
url = 'https://spaceimages-mars.com/'
browser.visit(url)
```

```python
image_link = browser.find_by_tag('button')[1]
image_link.click()
```

```python
html = browser.html
soup_image = BeautifulSoup(html, 'html.parser')

print(soup_image.prettify())
```

```python
image_url_find = soup_image.find('img',class_='fancybox-image').get('src')
image_url_find
```

```python
image_url = f'https://spaceimages-mars.com/{image_url_find}'
image_url
```

```python
url = 'https://marshemispheres.com/'
browser.visit(url)
```

```python
hemisphere_image_url = []

links = browser.find_by_css('a.product-item img')

for i in range(len(links)):
    
    hemisphereInfo={}
    
    browser.find_by_css('a.product-item img')[i].click
    
    sample = browser.links.find_by_text('Sample').first
    
    hemisphereInfo['img_url'] = sample['href']
    hemisphereInfo['title'] = browser.find_by_css('h2.title').text
    hemisphere_image_url.append(hemisphereInfo)
    browser.back()
```
